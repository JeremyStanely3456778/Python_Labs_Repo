Collecting Paragraph Portion of Webpage

```python
# import module
import requests
import pandas as pd
from bs4 import BeautifulSoup
  
# link for extract html data
def getdata(url):
    r = requests.get(url)
    return r.text
  
htmldata = getdata('https://en.wikipedia.org/wiki/Africa')
soup = BeautifulSoup(htmldata, 'html.parser')
data = ''
for data in soup.find_all("p"):
    print(data.get_text())
```

Collecting Image Filepaths and Images

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from os.path  import basename
import requests

  
htmldata = urlopen('https://en.wikipedia.org/wiki/Africa')
soup = BeautifulSoup(htmldata, 'html.parser')
images = soup.find_all('img')
print(images)
  
for item in images:
    print(item['src'])

i = 0  
for link in images:
    if "https//:" in link.get('src'):
        lnk = link.get('src')
        with open(basename(lnk), "wb") as f:
            f.write(requests.get(lnk).content)
    i += 1

print('\nThis total number image filepaths: ', i)
```

```python
from bs4 import BeautifulSoup
import requests

URL = "https://en.wikipedia.org/wiki/Africa"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
for a_href in soup.find_all("a", href=True):
    print(URL + a_href["href"])

archive_links = ["https://en.wikipedia.org/wiki/Africa"]

#save associated links to text file
for link in archive_links:
    page = requests.get(link)
    soup = BeautifulSoup(page.content, "html.parser")
    for a_href in soup.find_all("a", href=True):
        with open("wikipedia_africa_links.txt", "a") as linkfile:
            linkfile.write(a_href["href"]+"\n")
```

Collecting Tabular Data from Webpage

```python
import csv
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen(
    'https://en.wikipedia.org/wiki/Africa')
bs = BeautifulSoup(html, 'html.parser')

table = bs.findAll('table', {'class': 'wikitable'})[0]
rows = table.findAll('tr')

csvFile = open('Table_data_wikipedia_africa_page.csv', 'wt+')
writer = csv.writer(csvFile)
try:
    for row in rows:
        csvRow = []
        for cell in row.findAll(['td', 'th']):
            csvRow.append(cell.get_text())
            writer.writerow(csvRow)
            print(row.get_text())
finally:
    csvFile.close()
```
